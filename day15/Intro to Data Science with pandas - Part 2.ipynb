## Cleaning up our data

We're going to rename those impractical column names, and deal with the missing values in this data.

```python
# First let's rename those unusual column names:
movies.columns
```

```python
movies.rename(columns={
  'Runtime (Minutes)': 'Runtime',
  'Revenue (Millions)': 'Revenue_millions'
    
}, inplace=True)  # inplace means actually change the original movies table, instead of returning a new table

movies
```

```python
# Change all the column names to lowercase
# movies.columns = ['rank', 'genre', 'description', 'director'] # etc......

# list comprehension to the rescue!
movies.columns = [ col.lower() for col in movies.columns ]

movies.columns
```

### Dealing with missing values

```python
# Find the rows with null values
movies.isnull()
```

```python
# The above is a good start but too verbose, it's an entire new DataFrame
# of 1000 rows x 11 columns
# Let's just get a total of the null values in each column by using .sum()
movies.isnull().sum()
```

```python
# Ok, so the null values are all in the revenue_millions and metascore columns ('Series')
# But how to fill them in? We don't want to throw out the whole movie with the null
# value just because of one column....

# We can "impute" the values for those null column rows, i.e. replace them with a good-enough
# placeholder
# The most common choice is the mean (average) for that column

revenue_mean = movies['revenue_millions'].mean()
revenue_mean
```

```python
# Use .fillna() to replace the values
movies['revenue_millions'].fillna(revenue_mean, inplace=True)
# Check that it worked:
movies.isnull().sum()
```

```python
# ^ We could use a similar process to replace the nulls in the 'metascore'
#
# In practice, you might want to be a bit more refined about the values you use for your
# imputation... for example, replacing the revenue null with a revenue mean for all movies
# in the same genre as the movie you're replacing the null for (instead of just the mean
# revenue for all movies in the dataset).
```

```python
# To do conditional querying of the values in a column:
movies['director'] == 'Ridley Scott'
```

```python
# ^ This just gets us a True/False list (Series) for each, the result of the conditional
# test when applied to each row.
# To actually get the entire row for every film directed by Ridley Scott, we need to use
# the conditional test inside the movies[] brackets

just_ridley = movies[ movies['director'] == 'Ridley Scott' ]

# i.e. "Return all rows which have 'Ridley Scott' as the value of their 'director' column

just_ridley['revenue_millions'].mean()
```

```python
movies['revenue_millions'].mean()
```

```python
# What about finding the movies with only high ratings?

movies[ movies['rating'] >= 8.0 ]
```

```python
# ^ This result is not sorted by rating. To do that:

movies[ movies['rating'] >= 8.0 ].sort_values('rating', ascending=False).head(10)
```

```python
# Finding movies which feature a specific actor:

# movies[ 'Christian Bale' in movies['actors'] ]  # Whoops!!

movies[ movies['actors'].str.contains('Christian Bale') ]
```

```python
# Conditions on multiple columns: 
# i.e. show me movies directed by EITHER Ridley Scott OR Christopher Nolan

movies[ (movies['director'] == 'Ridley Scott') | (movies['director'] == 'Christopher Nolan') ]
```

```python
# A shorter way to write the above:
movies[ movies['director'].isin(['Christopher Nolan', 'Ridley Scott']) ]
```

```python
# A more complex query example:
# "Show me all movies released between 2005 and 2010, with a rating above 8.0,
# but whose revenue was below the 25th percentile for the whole dataset."
movies[
    ( (movies['year'] >= 2005) & (movies['year'] <= 2010) )
    & (movies['rating'] > 8.0)
    & (movies['revenue_millions'] < movies['revenue_millions'].quantile(0.25))
]
```

## Applying functions to every row in a Series or DataFrame

```python
# Let's create a new column for every movie, with a simple "good" or "bad" 
# value which is based on the current "rating" column's value for that movie

def number_rating_to_word( r ):
    if r >= 8.0:
        return 'good'
    else:
        return 'bad'

    # return 'good' if r >= 8.0 else 'bad'

    
number_rating_to_word( 7.0 )

# Now we want to call this function (apply it to) every rating value in our movies DataFrame.
# You could use a 'for' loop to do this, but's it much slower than using .apply() because
# of 'vectorization' - applying an operation to a whole list/array/column at once

good_or_bad = movies['rating'].apply( number_rating_to_word )

# Since the result here is a new Series (a new column) with a value for every movie row
# in the original DataFrame, we can just set this new column as a column of 'movies'

movies['rating_category'] = good_or_bad

movies.head()
```

```python
# For the brave:
# instead of defining a new function with its own name, we could a lambda function
# (anonymous function) as the argument to .apply()

movies['revenue_millions'].apply( lambda x: x/2 ) 
#movies['revenue_millions'].apply( lambda r: 'good' if r > 100.0 else 'bad' ) 
```
